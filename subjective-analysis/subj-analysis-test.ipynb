```python
interp_data = [interp_bus,interp_king,interp_mushroom,interp_stone]
bicubic = []
bilinear = []
lanczos3 = []
ogniewski = []

for x in interp_data:
    bicubic.append(x[0])
    bilinear.append(x[1])
    lanczos3.append(x[2])
    ogniewski.append(x[3])
```

```python
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
xaxis = ['bus','king','mushroom','stone']
xi = list(range(len(x)))
plt.xticks(xi,xaxis)


ax.bar(X - 0.3, bilinear, color = 'b', width = 0.2, align='center', label='bilinear')
ax.bar(X - 0.10, bicubic, color = 'g', width = 0.2, align='center', label='bicubic')
ax.bar(X + 0.1, lanczos3, color = 'r', width = 0.2, align='center', label='lanczos3')
ax.bar(X + 0.3, ogniewski, color = 'y', width = 0.2, align='center', label='Ogniewski')
ax.legend(loc='upper right')
```
