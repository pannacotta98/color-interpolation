```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
import numpy as np

with open("results_april_28.csv") as f:
  csv_string = f.read()

# Split the two tables
table1_str = csv_string[:csv_string.find("imageA")]
table2_str = csv_string[csv_string.find("imageA"):]

table1 = pd.read_table(StringIO(table1_str), sep=",")
table2 = pd.read_table(StringIO(table2_str), sep=",")
```

# Extract only the image results

```python
imageRes = table1.iloc[:,5:]
imageRes.head(n=3)
```

```python
imageSum = imageRes.sum(axis = 0, skipna=True)
imageSum
```

```python
color_snail = imageSum.iloc[0:3]
print(color_snail)
interp_bird = imageSum.iloc[3:7]
print(interp_bird)
interp_bus = imageSum.iloc[7:11]
print(interp_bus)
interp_king = imageSum.iloc[11:15]
print(interp_king)
interp_mushroom = imageSum.iloc[15:19]
print(interp_mushroom)
color_king = imageSum.iloc[19:22]
print(color_king)
```

### Bird image results does not lie in same order as the other images so append them seperately

```python
interp_data = [interp_bus,interp_king,interp_mushroom]
bicubic = []
bilinear = []
lanczos3 = []
ogniewski = []

for x in interp_data:
    bicubic.append(x[0])
    bilinear.append(x[1])
    lanczos3.append(x[2])
    ogniewski.append(x[3])

# Append bird image seperately
bilinear.append(interp_bird[0])
lanczos3.append(interp_bird[1])
ogniewski.append(interp_bird[2])
bicubic.append(interp_bird[3])
```

```python
color_data = [color_king, color_snail]
srgb = []
lab = []
linrgb = []

for x in color_data:
    lab.append(x[0])
    linrgb.append(x[1])
    srgb.append(x[2])
```

### Testa att plota medelvÃ¤rdet av poÃ¤ngen

```python
avg_bilinear = [x / table1.shape[0] for x in bilinear]
avg_bicubic = [x / table1.shape[0] for x in bicubic]
avg_lanczos3 = [x / table1.shape[0] for x in lanczos3]
avg_ogniewski = [x / table1.shape[0] for x in ogniewski]

avg_lab = [x / table1.shape[0] for x in lab]
avg_linrgb = [x / table1.shape[0] for x in linrgb]
avg_srgb = [x / table1.shape[0] for x in srgb]
```

# Plot the sum of the interpolation data

```python
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
xaxis = ['bus','king','mushroom','bird']
xi = list(range(len(xaxis)))
plt.xticks(xi,xaxis)
plt.rc('font', size=13)


ax.bar(X - 0.3, avg_bilinear, color = 'b', width = 0.2, align='center', label='bilinear')
ax.bar(X - 0.10, avg_bicubic, color = 'g', width = 0.2, align='center', label='bicubic')
ax.bar(X + 0.1, avg_lanczos3, color = 'r', width = 0.2, align='center', label='lanczos3')
ax.bar(X + 0.3, avg_ogniewski, color = 'y', width = 0.2, align='center', label='Ogniewski')
ax.legend(loc='upper left')
ax.set_ylim(0,3)
ax.set_ylabel('Average score')
```

# Plot the sum of the color space data

```python
X = np.arange(2)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
xaxis = ['king','snail']
xi = list(range(len(xaxis)))
plt.xticks(xi,xaxis)


ax.bar(X - 0.2, avg_lab, color = 'b', width = 0.2, align='center', label='CIELAB')
ax.bar(X - 0.0, avg_linrgb, color = 'g', width = 0.2, align='center', label='linRGB')
ax.bar(X + 0.2, avg_srgb, color = 'r', width = 0.2, align='center', label='sRGB')
ax.legend(loc='upper left')
ax.set_ylim(0,2)
ax.set_ylabel('Average score')
```
