```python
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
import numpy as np

with open("results_april_20.csv") as f:
  csv_string = f.read()

# Split the two tables
table1_str = csv_string[:csv_string.find("imageA")]
table2_str = csv_string[csv_string.find("imageA"):]

table1 = pd.read_table(StringIO(table1_str), sep=",")
table2 = pd.read_table(StringIO(table2_str), sep=",")
```

# Extract only the image results

```python
imageRes = table1.iloc[:,5:]
imageRes.head(n=3)
```

```python
imageSum = imageRes.sum(axis = 0, skipna=True)
imageSum
```

```python
color_snail = imageSum.iloc[0:3]
print(color_snail)
interp_bird = imageSum.iloc[3:7]
print(interp_bird)
interp_bus = imageSum.iloc[7:11]
print(interp_bus)
interp_king = imageSum.iloc[11:15]
print(interp_king)
interp_mushroom = imageSum.iloc[15:19]
print(interp_mushroom)
color_king = imageSum.iloc[19:22]
print(color_king)
```

### Bird image results does not lie in same order as the other images so append them seperately

```python
interp_data = [interp_bus,interp_king,interp_mushroom]
bicubic = []
bilinear = []
lanczos3 = []
ogniewski = []

for x in interp_data:
    bicubic.append(x[0])
    bilinear.append(x[1])
    lanczos3.append(x[2])
    ogniewski.append(x[3])

# Append bird image seperately
bilinear.append(interp_bird[0])
lanczos3.append(interp_bird[1])
ogniewski.append(interp_bird[2])
bicubic.append(interp_bird[3])
```

```python
color_data = [color_king, color_snail]
srgb = []
lab = []
linrgb = []

for x in color_data:
    lab.append(x[0])
    linrgb.append(x[1])
    srgb.append(x[2])
```

# Plot the sum of the interpolation data

```python
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
xaxis = ['bus','king','mushroom','bird']
xi = list(range(len(xaxis)))
plt.xticks(xi,xaxis)


ax.bar(X - 0.3, bilinear, color = 'b', width = 0.2, align='center', label='bilinear')
ax.bar(X - 0.10, bicubic, color = 'g', width = 0.2, align='center', label='bicubic')
ax.bar(X + 0.1, lanczos3, color = 'r', width = 0.2, align='center', label='lanczos3')
ax.bar(X + 0.3, ogniewski, color = 'y', width = 0.2, align='center', label='Ogniewski')
ax.legend(loc='upper right')
```

# Plot the sum of the color space data

```python
X = np.arange(2)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
xaxis = ['king','snail']
xi = list(range(len(xaxis)))
plt.xticks(xi,xaxis)


ax.bar(X - 0.2, lab, color = 'b', width = 0.2, align='center', label='CIELAB')
ax.bar(X - 0.0, linrgb, color = 'g', width = 0.2, align='center', label='linRGB')
ax.bar(X + 0.2, srgb, color = 'r', width = 0.2, align='center', label='sRGB')
ax.legend(loc='upper right')
```
